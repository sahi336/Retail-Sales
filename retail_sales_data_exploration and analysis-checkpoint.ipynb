```python
#connect to sql database
import sqlalchemy
```

```python
# Database connection
engine = sqlalchemy.create_engine('postgresql://postgres:password@localhost/retail_sales')
query = "SELECT * FROM sales"
retail_sales_df = pd.read_sql_query(query, engine)
```

```python
# Display the result
print(retail_sales_df.head())
```

```python
#top products by revenue
top_products = retail_sales_df.groupby('product_category')['total_amount'].sum().nlargest()
top_products.plot(kind='bar', title='Top Products by Revenue')
```

```python
#monthly sales trend
retail_sales_df['month'] = pd.to_datetime(retail_sales_df['date']).dt.to_period('M')
monthly_sales = retail_sales_df.groupby('month')['total_amount'].sum()
monthly_sales.plot(kind='line', title='Monthly Sales Trend')
```
